```python
def split_consecutive(arr):
    result = []
    current_group = [arr[0]]  # start with first element

    for i in range(1, len(arr)):
        if arr[i] == arr[i - 1]:
            current_group.append(arr[i])
        else:
            result.append(current_group)
            current_group = [arr[i]]

    result.append(current_group)  # add the last group
    return result


# Example
arr = [1,1,1,2,2,3,3,2,2,1]
print(split_consecutive(arr))
```
