# Covariance
Covariance is a measure of how much two random variables change together. 
It indicates the direction of the linear relationship between the variables.

A positive covariance means that the variables tend to increase or decrease together, while a negative covariance means that they tend to move in opposite directions. 

The magnitude of the covariance indicates the strength of the relationship, but it is not standardized and can be difficult to interpret without context.

```python
#calculate covariance of two variables using numpy
#
import numpy as np
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
covariance = np.cov(x, y)[0][1]
print("Covariance:", covariance)
print(np.cov(x, y))
```

```python
#covariance of two variables without using numpy
def covariance(x, y):
    n = len(x)
    mean_x = sum(x) / n
    mean_y = sum(y) / n
    cov = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n)) / (n - 1)
    return cov
y = [2, 3, 4, 5, 6]
x = [1, 2, 3, 4, 5]
cov = covariance(x, y)
print("Covariance:", cov)
#what does covariance of 2.5 mean and by how much do x and y increase 

```

# Correlation

Correlation is a standardized measure of the linear relationship between two variables, ranging from -1 to 1. Unlike covariance, correlation is unitless and easier to interpret.

- **Correlation = 1**: Perfect positive linear relationship
- **Correlation = -1**: Perfect negative linear relationship
- **Correlation = 0**: No linear relationship

```python
#correlation of two variables without using numpy
def correlation(x, y):  
    n = len(x)
    mean_x = sum(x) / n
    mean_y = sum(y) / n
    cov = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n)) / (n - 1)
    std_x = (sum((x[i] - mean_x) ** 2 for i in range(n)) / (n - 1)) ** 0.5
    std_y = (sum((y[i] - mean_y) ** 2 for i in range(n)) / (n - 1)) ** 0.5
    corr = cov / (std_x * std_y)
    return corr 
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
corr = correlation(x, y)
print("Correlation:", corr)
```

```python
#correlation of two variables using numpy
import numpy as np
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
corr = np.corrcoef(x, y)[0, 1]
print("Correlation:", corr) 
```
