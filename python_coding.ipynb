```python
lst=[2,1,6,3,7,5,4]
def second_large(lst):
    largest=lst[0]#assume first element as largest
    seclarge=lst[1]#assume second element as secondlarge
    for element in lst:
        if element>largest:
            largest=element
            seclarge=largest
        elif seclarge<largest and element=<seclarge:
            return seclarge
            
```

```python
lst=[1,2,3,4,1,2]
def dup(lst):
    unique=[]
    dupl=[]
    for element in lst:
        if element not in unique:
            unique.append(element)
        elif element not in dupl:
            dupl.append(element)
    return unique,dupl
print(dup(lst))    
```
